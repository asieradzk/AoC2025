# DDD

```polyglot-notebook
type Direction =
    | L
    | R

type Turn = Direction * int

type AllTurns = Turn list
```

```polyglot-notebook
let parseInput (input: string) : AllTurns =
    input.Split('\n')
    |> Array.toList
    |> List.map (fun line ->
        let dirChar = line.[0]
        let distanceStr = line.[1..]
        let direction =
            match dirChar with
            | 'L' -> L
            | 'R' -> R
            | _ -> failwithf "Invalid direction character: %c" dirChar
        let distance = int distanceStr
        (direction, distance)
    )
```

```polyglot-notebook
let countZeroPoints (turns: AllTurns) acc: int =
    let maxNumber = 99
    let startNumber = 50
    let rec loop turns acc position =
        match turns with
        | [] -> acc
        | (direction, distance) :: tail ->
            let newPosition =
                match direction with
                | L -> (position - distance + maxNumber + 1) % (maxNumber + 1)
                | R -> (position + distance) % (maxNumber + 1)
            let newAcc = if newPosition = 0 then acc + 1 else acc
            loop tail newAcc newPosition
    loop turns acc startNumber
```

```polyglot-notebook
let solution =
    let turns = parseInput testInput
    countZeroPoints turns 0

printfn "Solution: %d" solution
```

# pt2

```polyglot-notebook
let countAnyTimePointZero (turns: AllTurns) : int =
      let maxNumber = 99
      let startNumber = 50
      let rec loop turns position count =
          match turns with
          | [] -> count
          | (direction, distance) :: tail ->
              let rec moveSteps steps pos acc =
                  if steps = 0 then
                      loop tail pos acc
                  else
                      let newPos =
                          match direction with
                          | L -> (pos - 1 + maxNumber + 1) % (maxNumber + 1)
                          | R -> (pos + 1) % (maxNumber + 1)
                      let newAcc = if newPos = 0 then acc + 1 else acc
                      moveSteps (steps - 1) newPos newAcc
              moveSteps distance position count
      loop turns startNumber 0
```

```polyglot-notebook
let solutionpt2 =
    let turns = parseInput testInput
    countAnyTimePointZero turns

printfn "Solution pt2: %d" solutionpt2
```
